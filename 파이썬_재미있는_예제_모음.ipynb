# 야구게임

```
import random
numbers = list("1234567890")
random.shuffle(numbers)
answers = numbers[:3]

strike = 0
while strike < 3:
    inputs = input("세자리 숫자를 입력하세요\n")
    strike = sum([c1 == c2 for c1, c2 in zip(inputs, answers)])
    ball = sum([c in answers for c in inputs]) - strike
    print(f"{strike} strike, {ball} ball")

print('정답입니다.')
```

```
# list comprehension
word_count = [(w, words.count(w)) for w in set(words) if w != '']
word_sorted = sorted(word_count, key=lambda x: x[1], reverse=True)
word_sorted[:10]
```

```
import re
 
curr = '1'
for each in range(30):
    match = re.findall("(\d)(\\1*)", curr)
    print(curr)
    print(match)
    curr = ''.join([str(len(k + y)) + k for k, y in match])

print(len(curr))
```
