## 03. Make a Graph with TF

```python
'''
print(X_train.shape, Y_train.shape)
(1600, 1),   (1600, 1)
'''
# Graph Clear & Make your Graph reproducible
tf.reset_default_graph()
tf.set_random_seed(2017)

# Hyper-Parameters & Option
learning_rate = 0.01

# Placeholders
X = tf.placeholder(dtype=tf.float32, shape=[None, 1])
Y = tf.placeholder(dtype=tf.float32, shape=[None, 1])

# Weight & Bias
W = tf.Variable(tf.random_normal([1, 1]))
b = tf.Variable(tf.random_normal([1]))

# Hypothesis
Y_pred = tf.add(tf.matmul(X, W), b)

# Cost(loss) function & Optimizer
cost = tf.reduce_mean(tf.squared_difference(Y, Y_pred))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

## 04. Train! Session!

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

learning_rate = 1

start = time.clock()
for i in range(2000):
    # 학습 진행
    sess.run(optimizer, feed_dict={X: X_train, Y: Y_train})
    
    if (i % 100 == 0):
        # 학습 상황 디스플레이
        tr_l = sess.run(cost, feed_dict={X: X_train, Y: Y_train})
        print("{}th training---- training loss: {:.2f}".format(i, tr_l))

print("Training Complete")
end = time.clock()

print("Total elapsed time: {}".format(end - start))
print("Test cost(loss): {:.2f}".format(sess.run(cost, feed_dict={X: X_test, Y: Y_test})))

Y_predicted = sess.run(Y_pred, feed_dict={X: X_test, Y: Y_test})
```

#### 데이터의 인풋은 온도였습니다. 아웃풋은 판매량이었구요.

```python
print(sess.run([W, b]))
```

```python
# 세션을 닫아야 합니다.
sess.close()
```

### Dummy Variable & Drop

데이터셋 안에 '국가'라는 컬럼이 있다고 생각해보자.<br>
그걸 그대로 사용할 때 펼쳐질 에러에 대해서 논해보자.

```python
# 더비 변수 생성, one hot encoding
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)

# 필요없는 칼럼 제거
fields_to_drop = ['instant', 'dteday', 'season', 'weathersit', 'mnth', 'hr', 'weekday']
data = rides.drop(fields_to_drop, axis=1)
data.head()
```

### Scaling
결과값 검증할 때에는 scailing했던 내용을 반대로 적용(**unscailing**)해서 검증해야 한다. <br>
아래 쪽에 **unscaling**을 위한 코드 부분이 나올 때에 비교해서 살펴보도록 하자.

```python
# scaling 이전의 데이터
data[['casual', 'registered', 'cnt', 'temp', 'atemp', 'hum', 'windspeed']].head(n=10)
```

```python
quant_features = ['casual', 'registered', 'cnt', 'temp', 'atemp', 'hum', 'windspeed']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean) / std
    
data[['casual', 'registered', 'cnt', 'temp', 'atemp', 'hum', 'windspeed']].head(n=10)
```

### 실습

```python
import pandas as pd 

# 1. 'https://raw.githubusercontent.com/blackdew/dplus-tensorflow/master/csv/Bike-Sharing-Hour.csv' 데이터를 읽어들인다.
rides = pd.read_csv('https://raw.githubusercontent.com/blackdew/dplus-tensorflow/master/csv/Bike-Sharing-Hour.csv')

# 2. ['season', 'weathersit', 'mnth', 'hr', 'weekday'] 칼럼은 더미변수화 한다.
dummy_fields = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for each in dummy_fields:
    dummies = pd.get_dummies(rides[each], prefix=each, drop_first=False)
    rides = pd.concat([rides, dummies], axis=1)
    
# 3. 필요 없어진 ['instant', 'dteday', 'season', 'weathersit', 'mnth', 'hr', 'weekday'] 칼럼을 제거한다.
fields_to_drop = ['instant', 'dteday', 'season', 'weathersit', 'mnth', 'hr', 'weekday']
data = rides.drop(fields_to_drop, axis=1)
data.head()

# 4. ['casual', 'registered', 'cnt', 'temp', 'atemp', 'hum', 'windspeed'] 칼럼을 scailing 한다. 
quant_features = ['casual', 'registered', 'cnt', 'temp', 'atemp', 'hum', 'windspeed']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean) / std
    
# 5. test set과 training set로 분리한다.
train_data, test_data = data[:-61*24], data[-61*24:]

target_fields = ['cnt', 'casual', 'registered']
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
train_features, train_targets = train_data.drop(target_fields, axis=1), train_data[target_fields]

# 6. ['cnt', 'casual', 'registered'] 칼럼의 값을 target value로 지정한다.
test_X, test_Y = test_features.values, test_targets.values
train_X, train_Y = train_features.values, train_targets.values
```

```python
'''
TensorFlow 그래프를 전체를 직접 작성해 보자
'''
import tensorflow as tf 

# Graph Clear & Make your Graph reproducible
tf.reset_default_graph()
tf.set_random_seed(2017)

# Hyper-Parameters & Option
learning_rate = 0.01

# Placeholders
X = tf.placeholder(dtype=tf.float32, shape=[None, 58])
Y = tf.placeholder(dtype=tf.float32, shape=[None, 3])

# Weight & Bias
W = tf.Variable(tf.random_normal([58, 3]))
b = tf.Variable(tf.random_normal([3]))

# Hypothesis
Y_pred = tf.add(tf.matmul(X, W), b)

# Cost(loss) function & Optimizer
cost = tf.reduce_mean(tf.squared_difference(Y, Y_pred))
Optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
'''
학습하는 세션 코드 전체를 직접 작성해 보자
'''
sess = tf.Session()
sess.run(tf.global_variables_initializer())

learning_rate = 0.01

for i in range(1000):
    # 학습 진행
    sess.run(Optimizer, feed_dict={X: train_X, Y: train_Y})
    
    if (i % 100 == 0):
        # 학습 상황 디스플레이
        tr_l = sess.run(cost, feed_dict={X: train_X, Y: train_Y})
        print("{}th training---- training loss: {:.2f}".format(i, tr_l))
        
print("Training Complete")

print("Test loss: {:.2f}".format(sess.run(cost, feed_dict={X: test_X, Y: test_Y})))
```
