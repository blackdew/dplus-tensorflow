### 03.5 실습 - Make a CNN Graph with TF

```python
'''
버그를 해결해 주세요. 
잘못된 곳은 6군데 입니다.
'''

# Graph Clear
tf.reset_default_graph()
tf.set_random_seed(2017) # random seeding - reproduct

X = tf.placeholder(tf.float32, [None, 28 * 28], name="X")
X_img = tf.reshape(X, [-1, 28, 28, 1], name="X_img") 
Y = tf.placeholder(tf.float32, [None, 10], name="Y")

# for Conv Layer 01 filter
# W1 = tf.Variable(tf.random_normal([3, 3, 3, 16], stddev=0.01))
W1 = tf.Variable(tf.random_normal([3, 3, 1, 32], stddev=0.01))
conv1 = tf.nn.conv2d(X_img, W1, strides=[1, 1, 1, 1], padding='SAME')
conv1 = tf.nn.relu(conv1)
pool1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# for Conv Layer 02 filter
W2 = tf.Variable(tf.random_normal([3, 3, 32, 64], stddev=0.01))
conv2 = tf.nn.conv2d(pool1, W2, strides=[1, 1, 1, 1], padding='SAME')
conv2 = tf.nn.relu(conv2)
pool2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# fully connected layer
# flat = tf.reshape(pool2, [-1, 8 * 8 * 64])
# wfc = tf.Variable(tf.random_normal([8 * 8 * 64, 20], stddev=0.01))
flat = tf.reshape(pool2, [-1, 7 * 7 * 64])
wfc = tf.Variable(tf.random_normal([7 * 7 * 64, 10], stddev=0.01))
bfc = tf.Variable(tf.random_normal([10]))
logits = tf.add(tf.matmul(flat, wfc), bfc)

# Cost(loss) function - softmax
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=Y), name="cost")

# Optimizer - Adam
learning_rate = 0.001
# optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize()
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

## 05. 실습 해답

### 05.1 Import Library & Prepare MNIST Data

```python
import tensorflow as tf
import numpy as np

mnist = tf.contrib.learn.datasets.load_dataset("mnist")
```

### 05.2 Prepare Variables for a CNN Graph with TF

```python
# Graph Clear
tf.reset_default_graph()
tf.set_random_seed(2017) # random seeding - reproduct

############################
# Place Holders

X = tf.placeholder(tf.float32, [None, 784])
X_img = tf.reshape(X, [-1, 28, 28, 1])
Y = tf.placeholder(tf.float32, [None, 10])
```

### 05.3 Make a CNN Graph & Run Session with TF

```python
#########################
# ConvLayer 01 

with tf.name_scope("layer1_conv"):

    # for Conv Layer 01 filter - shape=(3, 3, 1)
    W1 = tf.Variable(tf.random_normal([3, 3, 1, 24], stddev=0.01))

    # Convolution Layer 01 -> (?, 28, 28, 32)
    conv1 = tf.nn.conv2d(X_img, W1, strides=[1, 1, 1, 1], padding='SAME')
    conv1 = tf.nn.relu(conv1)

    # Pooling Layer 01 -> (?, 14, 14, 32)
    pool1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

    
#########################
# ConvLayer 02

with tf.name_scope("layer2_conv"):

    # for Conv Layer 02 filter - shape=(3, 3, 32)
    W2 = tf.Variable(tf.random_normal([3, 3, 24, 64], stddev=0.01))

    # Convolution Layer 02 -> (?, 14, 14, 32)
    conv2 = tf.nn.conv2d(pool1, W2, strides=[1, 1, 1, 1], padding='SAME')
    conv2 = tf.nn.relu(conv2)

    # Pooling Layer 02 -> (?, 7, 7, 64)
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

    
#########################
# Flatten layer & Fully Connected Layer

with tf.name_scope("L_FC"):

    ###############################
    # Flatten layer
    flat = tf.reshape(conv2, [-1, 7 * 7 * 64])

    ###############################
    # Fully Connected Layer

    # for Final FC 7x7x64 inputs -> 10 outputs
    wfc = tf.Variable(tf.random_normal([7 * 7 * 64, 10], stddev=0.01))

    # bias
    bfc = tf.Variable(tf.random_normal([10]))
    
    # fc layer
    logits = tf.add(tf.matmul(flat, wfc), bfc)

    
#########################
# Cost & Optimizer

learning_rate = 0.001

with tf.name_scope("Optimizer"):
    # Cost(loss) function & Optimizer
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=Y))
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

### 05.4 Training & Evaluation

```python
#########################
# Session initialize

sess = tf.Session()
sess.run(tf.global_variables_initializer())


###########################
# Traning

print('Learning started. It takes sometime.')

# 55000개의 데이터 500개씩 110번 가져와야 전체를 학습하게 된다.
epochs = 10
batch_size = 500
n_of_batches = int(mnist.train.num_examples / batch_size)

# epochs회 학습 
for e in range(epochs):
    print("ephoch: %dth" % (e + 1))

    # n_of_batches회 동안 batch_size개씩 가져와서 학습 진행
    for i in range(n_of_batches):
        X_batch, Y_batch = mnist.train.next_batch(batch_size)
        sess.run(optimizer, feed_dict={X: X_batch, Y: np.eye(10)[Y_batch]})

        # 학습 상황 디스플레이
        if ((i + 1) % 10 == 0):
            loss = sess.run(cost, feed_dict={X: X_batch, Y: np.eye(10)[Y_batch]})
            print("training cost: %.3f" % loss)

            
print("Training Complete")

###########################
# Evaluation
with tf.name_scope("Prediction"):
    # Test model and check accuracy
    correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(Y, 1))
    acc = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    accuracy = sess.run(acc, feed_dict={X: mnist.test.images, Y: np.eye(10)[mnist.test.labels]})

print('Accuracy: %.2f' % (accuracy * 100))
```

## 06. 대체 뭘 학습한 것인가? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np

def plot_conv_weights(weights, input_channel=0):
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())

        w = sess.run(weights)

        w_min = np.min(w)
        w_max = np.max(w)

        num_filters = w.shape[3]
        num_grids = math.ceil(math.sqrt(num_filters))
        fig, axes = plt.subplots(num_grids, num_grids)

        for i, ax in enumerate(axes.flat):
            if i < num_filters:
                img = w[:, :, input_channel, i]
                ax.imshow(img, vmin=w_min, vmax=w_max, interpolation='nearest', cmap='seismic')

            ax.set_xticks([])
            ax.set_yticks([])

        plt.show()
```
