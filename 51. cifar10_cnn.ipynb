## 05. 실습 해답

### 05.1 Import Library & Prepare CIFAR-10 Data

```python
import tensorflow as tf
import numpy as np
import random

# cifar10
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']

(train_x, train_y), (test_x, test_y) = tf.keras.datasets.cifar10.load_data()

train_y = train_y.reshape(-1,)
test_y = test_y.reshape(-1,)

'''
트레이닝 데이터로 활용할 50000개 이미지, 테스트용 10000개 이미지
32 * 32 * 3 사이즈의 이미지가 어레이에 담겨있다.
레이블은 one-hot encoding이 되어 있지 않다.
'''

print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

### 05.2 Prepare Variables for a CNN Graph with TF

```python
# Graph Clear
tf.reset_default_graph()

############################
# Place Holders

X = tf.placeholder(tf.float32, [None, 32, 32, 3])
Y = tf.placeholder(tf.float32, [None, 10])

lr = tf.placeholder(tf.float32)
kp = tf.placeholder(tf.float32)
```

### 05.3 Make a CNN Graph & Run Session with TF

```python
#########################
# ConvLayer 01 

# for Conv Layer 01 filter
w1 = tf.get_variable(shape=[5, 5, 3, 16], initializer=tf.contrib.layers.xavier_initializer(), name="w1")
b1 = tf.Variable(tf.random_normal([16], stddev=0.01))

# Convolution Layer -> (?, 32, 32, 16)
conv1 = tf.add(tf.nn.conv2d(X, w1, strides=[1, 1, 1, 1], padding='SAME'), b1)
conv1 = tf.nn.relu(conv1)

# Pooling Layer -> (?, 16, 16, 16)
pool1 = tf.nn.max_pool(conv1, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')

#########################
# ConvLayer 02

# for Conv Layer 02 filter
w2 = tf.get_variable(shape=[3, 3, 16, 16], initializer=tf.contrib.layers.xavier_initializer(), name="w2")
b2 = tf.Variable(tf.random_normal([16], stddev=0.01))

# Convolution Layer -> (?, 16, 16, 16)
conv2 = tf.add(tf.nn.conv2d(pool1, w2, strides=[1, 1, 1, 1], padding='SAME'), b2)
conv2 = tf.nn.relu(conv2)

# Pooling Layer -> (?, 8, 8, 16)
pool2 = tf.nn.max_pool(conv2, ksize=[1, 3, 3, 1], strides=[1, 2, 2, 1], padding='SAME')
    
##################################################
# Flatten layer & Fully Connected Layer

###############################
# Flatten layer
flat = tf.reshape(pool2, [-1, 8 * 8 * 16])

###############################
# Fully Connected Layer 01

# for Final FC 8x8x64 inputs -> 256 outputs
wfc1 = tf.get_variable(shape=[8 * 8 * 16, 128], initializer=tf.contrib.layers.xavier_initializer(), name="wfc1")
bfc1 = tf.Variable(tf.random_normal([128], stddev=0.01))

# fc layer
fc1 = tf.add(tf.matmul(flat, wfc1), bfc1)
fc1 = tf.nn.relu(fc1)

# dropout
fc1 = tf.nn.dropout(fc1, keep_prob=kp)

###############################
# Fully Connected Layer 02

# for Final FC 256 inputs -> 10 outputs
wfo = tf.get_variable(shape=[128, 10], initializer=tf.contrib.layers.xavier_initializer(), name="wfo")
bfo = tf.Variable(tf.random_normal([10], stddev=0.01))

# fc layer
logits = tf.add(tf.matmul(fc1, wfo), bfo)

#########################
# Cost & Optimizer

# Cost(loss) function & Optimizer
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=Y))
optimizer = tf.train.AdamOptimizer(learning_rate=lr).minimize(cost)
    
# Test model and check accuracy
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(Y, 1))
acc = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

### 05.4 Training & Evaluation

```python
#########################
# Session initialize

sess = tf.Session()
sess.run(tf.global_variables_initializer())

###########################
# Traning

print('Learning started. It takes sometime.')

epochs = 20
batch_size = 256
n_of_batches = int(len(train_x) / batch_size)

for epoch in range(epochs): 
    print("%dth epoch" % (epoch + 1))

    for i in range(n_of_batches):
        X_batch = train_x[(i * batch_size):((i + 1) * batch_size)]
        Y_batch = train_y[(i * batch_size):((i + 1) * batch_size)]

        sess.run(optimizer, feed_dict={X: X_batch, Y: np.eye(10)[Y_batch], lr: 0.05, kp: 0.7})

        # 학습 상황 디스플레이
        if ((i + 1) % 20 == 0): 
            loss, accuracy = sess.run([cost, acc], feed_dict={X: X_batch, Y: np.eye(10)[Y_batch], lr: 0.05, kp: 1.0})
            print("%dth records, training cost: %.3f, accuracy: %.2f" % ((i + 1) * batch_size, loss, accuracy * 100))

print("Training Complete")
```

```python
###########################
# Evaluation
accuracy = sess.run(acc, feed_dict={X: test_x, Y: np.eye(10)[test_y], lr: 0.01, , kp: 1.0})
print('Accuracy: %.2f' % (accuracy * 100))
```

## 06. 대체 뭘 학습한 것인가? 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np

def plot_conv_weights(weights, input_channel=0):
    
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())

        w = sess.run(weights)

        w_min = np.min(w)
        w_max = np.max(w)

        num_filters = w.shape[3]
        num_grids = math.ceil(math.sqrt(num_filters))
        fig, axes = plt.subplots(num_grids, num_grids)

        for i, ax in enumerate(axes.flat):
            if i < num_filters:
                img = w[:, :, input_channel, i]
                ax.imshow(img, vmin=w_min, vmax=w_max, interpolation='nearest', cmap='seismic')

            ax.set_xticks([])
            ax.set_yticks([])

        plt.show()
        
def plot_conv_layer(layer, image):

    with tf.Session() as sess:

        sess.run(tf.global_variables_initializer())

        feed_dict = {X: [image]}
        values = sess.run(layer, feed_dict=feed_dict)

        num_filters = values.shape[3]
        num_grids = math.ceil(math.sqrt(num_filters))
        fig, axes = plt.subplots(num_grids, num_grids)

        for i, ax in enumerate(axes.flat):
            if i < num_filters:
                img = values[0, :, :, i]
                ax.imshow(img, interpolation='nearest', cmap='binary')
                
            ax.set_xticks([])
            ax.set_yticks([])

        plt.show()
```
